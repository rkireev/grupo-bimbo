```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
clients = pd.read_csv('cliente_tabla.csv')
```

```python
clients['OXXO'] = clients['NombreCliente'].str.contains('OXXO')
clients['ARTELI'] = clients['NombreCliente'].str.contains('ARTELI')
clients['ALSUPER'] = clients['NombreCliente'].str.contains('ALSUPER')
clients['BODEGA'] = clients['NombreCliente'].str.contains('BODEGA')
clients['CALIMAX'] = clients['NombreCliente'].str.contains('CALIMAX')
clients['XICANS'] = clients['NombreCliente'].str.contains('NEZ|JOSE|NZO|TES')
clients['ABARROTES'] = clients['NombreCliente'].str.contains('ABARROTES')
clients['CARNICERIA'] = clients['NombreCliente'].str.contains('CARNICERIA')
clients['FRUTERIA'] = clients['NombreCliente'].str.contains('FRUTERIA')
clients['DISTRIBUIDORA'] = clients['NombreCliente'].str.contains('DISTRIBUIDORA')
clients['ELEVEN'] = clients['NombreCliente'].str.contains('SEVEN|ELEVEN')
clients['HOTEL'] = clients['NombreCliente'].str.contains('HOTEL')
clients['HOSPITAL'] = clients['NombreCliente'].str.contains('HOSPITAL|CLINICA')
clients['CAFE'] = clients['NombreCliente'].str.contains('CAFE')
clients['FARMACIA'] = clients['NombreCliente'].str.contains('FARMACIA')
clients['CREME'] = clients['NombreCliente'].str.contains('CREME')
clients['SUPER'] = clients['NombreCliente'].str.contains('SUPER')
clients['COMOD'] = clients['NombreCliente'].str.contains('COMOD')
clients['MODELOR'] = clients['NombreCliente'].str.contains('MODELOR')
```

```python
clients[clients['NombreCliente'].str.contains('MODELOR')]
```

```python
clients['UNKN'] = ~clients['UNKN']
```

```python
clients['UNKN'] = clients['OXXO']|clients['ARTELI']|clients['ALSUPER']|clients['BODEGA']|clients['CALIMAX']|clients['XICANS']|clients['ABARROTES']|clients['CARNICERIA']|clients['FRUTERIA']|clients['DISTRIBUIDORA']|clients['ELEVEN']|clients['HOTEL']|clients['HOSPITAL']|clients['CAFE']|clients['FARMACIA']|clients['CREME']|clients['SUPER']|clients['COMOD']|clients['MODELOR']
```

```python
del clients['NombreCliente']
```

```python
clients = clients.astype(int)
```

```python
clients.to_csv('clients')
```
