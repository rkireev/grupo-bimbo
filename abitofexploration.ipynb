```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
train = pd.read_csv('train.csv', usecols=[u'Semana', 'Producto_ID', u'Demanda_uni_equil'])
```

```python
popularity = train[train['Semana']==9]['Producto_ID'].value_counts()
popularity.head()
```

```python
popularity[popularity==np.round(popularity.median())]
```

```python
plt.figure(figsize=(10,10))
a = train[train['Producto_ID']==2233].reset_index() #Pan Blanco
for w in range(3,10):
    plt.subplot(708+w)
    plt.hist(a[a['Semana']==w]['Demanda_uni_equil'], bins=range(0,20))
plt.show()
```

```python
prod_group = train.groupby(['Semana', 'Producto_ID'])['Demanda_uni_equil'].median().reset_index()
```

```python
for w in range(3,10):
    plt.plot(prod_group[prod_group['Semana']==w].index, 
                prod_group[prod_group['Semana']==w]['Demanda_uni_equil'], '*')

plt.show()
```

```python
a = prod_group[prod_group['Producto_ID']==2233]
for w in range(3,10):
    plt.plot(a[a['Semana']==w].index, 
                a[a['Semana']==w]['Demanda_uni_equil'], '*')

plt.show()
```
